```python
# Read in new csv
df = pd.read_csv("Resources/2020-citibike-all.csv")
df.head()
```

```python
# Drop rows where the trip duration is less than 90 seconds and where the start and end stations are the same
df = df.loc[((df["tripduration"]>90) & (df["start station id"] != df["end station id"])),:]
df.head()
```

```python
# Convert trip duration column from seconds to minutes
df["tripduration"] = [trip/60 for trip in df["tripduration"]]

# Filter out trips over 24 hours long
df = df.loc[df["tripduration"] < 1440, :]

df.rename(columns={"tripduration":"tripduration (min)"})
df.head()
```

```python

# Rename values in gender column with strings
gender = {0:"Unknown", 1:"Male", 2:"Female"}
df.gender = [gender[item] for item in df.gender]
df.head()
```

```python
# Checking to see count of datapoints
df.count()
```

```python
# Dropping NaN
df = df.dropna(how='any')
df.count()
```

```python
# Save DataFrame to CSV
df.to_csv("Resources/2020-citibike-clean.csv", index=False, header=True)
```
